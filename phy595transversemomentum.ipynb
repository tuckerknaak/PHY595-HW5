#### Homework 2 - Problem 5.  We consider the decay of a $W^+$ boson into a positron $e^+$ and an electron neutrino $\nu_e$ through the weak force.  We can solve for the momentum of the positron $p_e^*$ in the rest frame of the boson in terms of the boson mass $m_W$.  We can then find the transverse component of the positron momentum $p_T$ in the lab frame in terms of $m_W$ and the angle $\theta^*$ with respect to the axis of motion of the boson.  From here, we can transform the isotropic probability density function $P(\theta^*)=\sin\theta^*$ from a function of $\theta^*$ to a function of $p_T$.  In this code, we plot the probability density $P(p_T)$ for the allowed values of the lab transverse momentum.

```python
'''Imports'''
import matplotlib.pyplot as plt
import numpy as np
np.seterr(divide='ignore');
```

```python
'''Mass of W+ Boson [Mev]'''
mass_W = 80370


'''Allowed Lab Transverse Momentum Values [Mev]'''
pt_points = np.linspace(0, 0.5 * mass_W, 1000)


'''Function to Calculate Probability Density of Lab Tranverse Momentum [1 / Mev]'''
def pt_density(pt):
    return (2 * pt / (mass_W * np.sqrt(mass_W**2 - 4 * pt**2)))


'''Probability Density of Lab Tranverse Momentum [1 / Mev]'''
pt_density_points = [pt_density(pt) for pt in pt_points]


'''Plotting the Probability Density'''
fig, ax = plt.subplots(figsize = (8, 5), dpi = 100)
ax.plot(pt_points, pt_density_points, linestyle = 'solid', color = 'red', 
        label = '$P(p_T)\ =\ 2p_T\ /\ m_W\sqrt{m_W^2-4p_T^2}$')
ax.grid(True, linestyle = 'dashed', color = 'darkgray', alpha = 0.5)
ax.set_title('$W^+ \u2192\  e^+ +\u03BD_e$')
ax.set_ylabel('Probability Density $P(p_T)$ [1 / Mev]')
ax.set_xlabel('Lab Tranverse Momentum $p_T$ [Mev]')
ax.set_xticks([0, mass_W * 0.0625, mass_W * 0.125, mass_W * 0.1875, mass_W * 0.25,
                  mass_W * 0.3125, mass_W * 0.375, mass_W * 0.4375, mass_W * 0.50],
              ['0', '$m_W\ /\ 16$', '$m_W\ /\ 8$', '$3m_W\ /\ 16$', '$m_W\ /\ 4$',
                    '$5m_W\ /\ 16$', '$3m_W\ /\ 8$', '$7m_W\ /\ 16$', '$m_W\ /\ 2$'])
ax.annotate('Jacobian Peak', (pt_points[-2], pt_density_points[-2]), textcoords = 'offset points',
            xytext = (-41, -10), ha = 'center')
ax.legend(loc = 'upper left', prop = {'size': 12}, shadow = True);
```
